```python
def calculate_mos(duration):
    
    # Split the string by whitespace
    parts = duration.split()
    years = 0
    months = 0
    
    # Getting the exact element of years and months
    if len(parts) == 4:
        years = int(parts[0])
        months = int(parts[2])
    elif len(parts) == 2:
        years = 0
        months = int(parts[0])

    total_mos = (years * 12) + months    
    return total_mos      
```
